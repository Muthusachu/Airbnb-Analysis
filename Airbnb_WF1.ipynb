**Collecting the Amenities**

```python
amenities= []
for i in collect.find({},{"_id":1, "amenities":1}):
    amenities.append(i)
```

```python
df_amts= pd.DataFrame(amenities)
```

```python
df_amts.head(2)
```

```python
def sort_amenities(x):
  a= x
  a.sort()
  return a
```

```python
df_amts["amenities"]= df_amts["amenities"].apply(lambda x: sort_amenities(x))
```
